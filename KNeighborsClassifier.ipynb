## irisデータセットを用いたKNeighborsClassifierの考察

```python
from algorithm import KNeighborsClassifier #今回実装したk点近傍法分類器
from sklearn.datasets import load_iris #irisデータセット
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris() #irisデータセットの準備

neighbors = list() #横軸
accuracy = list() #縦軸

for i in range(1,50):
    acc = KNeighborsClassifier.leave_one_out_accuracy( #一つ抜き法
        iris.data,
        iris.target,
        n_neighbors = i  #適用する近傍数を変動させる
    )  
    neighbors.append(i)
    accuracy.append(acc)
```

```python
'''
グラフの描画
'''
fig = plt.figure(figsize = (10,4))
ax = fig.add_subplot(111)
ax.plot(neighbors,accuracy,marker = "o")
ax.set_xlabel('n_neighbors')
ax.set_ylabel('leave one out accuracy')
plt.show()
```

```python
'''
最大値をとるn_neighborsの値を算出
'''

accuracy = np.array(accuracy)
neighbors = np.array(neighbors)
mask = accuracy == np.max(accuracy)
neighbors[mask]
```

### n_neighbors = 19,21 の時に最大値をとる。
