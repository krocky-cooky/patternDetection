```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
data = iris.data
```

```python
clf = KMeans(n_clusters = 4)
clf.fit(data)
for i in range(3):
    for j in range(i+1,4):
        clf.visualize(argx=i,argy=j)
        
```

```python
for i in range(0,2):
    for j in range(i+1,3):
        for t in range(j+1,4):
            clf.visualize3D(
                argx = i,
                argy = j,
                argz = t
            )
```
