## irisデータセットを用いたKMeansの考察

```python
from algorithm import KMeans #今回実装したk-means
from sklearn.datasets import load_iris #irisデータセット
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris() #irisデータセットの準備

clf = KMeans(n_clusters = 3)
clf.fit(iris.data) #データの入力
```

```python
'''
データの4つの要素のうち任意の二つにおけるクラスタ分類の可視化
'''

for i in range(0,3):
    for j in range(i+1,4):
        clf.visualize( #二次元可視化メソッド
            argx = i,
            argy = j
        )
```

```python
'''
データセットの4つの要素のうち任意の3つの要素を用いた3次元での可視化
'''

for i in range(0,2):
    for j in range(i+1,3):
        for t in range(j+1,4):
            clf.visualize3D( #3次元可視化メソッド
                argx = i,
                argy = j,
                argz = t
            )
```

### 各visualizeメソッドには引数targetを加えることで正解ラベルの可視化もできる。
以下の正解ラベルとクラスタ分類を比較すると、ある程度正しい分類ができていることがわかる。

```python
for i in range(0,3):
    for j in range(i+1,4):
        clf.visualize( #二次元可視化メソッド
            argx = i,
            argy = j,
            target = iris.target
        )
```

```python
for i in range(0,2):
    for j in range(i+1,3):
        for t in range(j+1,4):
            clf.visualize3D( #3次元可視化メソッド
                argx = i,
                argy = j,
                argz = t,
                target = iris.target
            )
```

```python
'''
n_clustersを2から5まで変化させてプロットする。
'''

for i in range(2,6):
    clf = KMeans(n_clusters = i)
    clf.fit(iris.data)
    clf.visualize3D(
        argx = 0,
        argy = 2,
        argz = 3
    )
```
